# Compare Different Strategy

### Calculate board expectation

```python
import CandyCrashSimulation3X3_2 as f
import random
import copy
import numpy as np
```

```python
width = height = 3
s = 2 ** (width * height)
num_candy = 2
N = 10000
print("size: {} * {}".format(width,height))
```

```python
w_score = np.zeros(s*N).reshape(s,N)
for i in range(s):
    board = f.buildboard(i, height, width)
    for c in range(N):
        newboard,tmp,score = f.eliminate_to_static(board,height,width)
        w_score[i][c] = score
mean = np.mean(w_score,axis=1)
std = np.std(w_score,axis=1)
with open('experiment_mean_var_width3_height3_candies2.txt','w') as fout:
    for i in range(s):
        fout.write("{} {} {}\n".format(i,mean[i],std[i]))
#print(mean)
```

## Random elimination or Based on expectation with 3X3 2 candy

```python
def compare(num_move):
    eachscore = np.zeros(3)
    eachscore[0] = num_move
    N = 1000
    for i in range(N):
        o_board = f.initialize(height,width)
        board = copy.deepcopy(o_board)
        #print("origin")
        #print(np.array(board))
        for j in range(num_move): # random move
            move = f.detect(board,height,width)
            ran = random.choice(move)
            board = f.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = f.eliminate_to_static(board,height,width)
            eachscore[1] += score
            #print("after")
            #print(np.array(board))
            #print('random',ran,str(score))
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # based expectation
            move = f.detect(board,height,width)
            i_move = 0
            max_value = -1 
            for i in move:                 # possible move
                tryboard = copy.deepcopy(board)
                tryboard = f.swap(tryboard,i//100,(i%100)//10,i%10)
                #print("tryboard")
                #print(np.array(tryboard))
                hashvalue = f.encode(tryboard,height,width)
                if mean[hashvalue] > max_value:
                    i_move = i
                    max_value = mean[hashvalue]
                    #print('change')
                #print('move',i,hashvalue,mean[hashvalue])
            #print("============")
            #print("origin")
            #print(np.array(board))
            board = f.swap(board,i_move//100,(i_move%100)//10,i_move%10)
            #print("after swap")
            #print(np.array(board))
            board,tmp,score = f.eliminate_to_static(board,height,width)
            #print("final")
            #print(np.array(board))
            #print('based',i_move,str(score))
            #print("============")
            eachscore[2] += score
            
    eachscore[1:] /= N
    return eachscore
```

```python
import csv
result = []
for i in range(1,50,1):
    tmp = compare(i)
    result.append(tmp)
    print(tmp)
with open('compare3X3.csv','w') as fout:
    writer = csv.writer(fout)
    writer.writerows(result)    
```

## Random elimination | Horizontal Elimination first | Vertical Elimination first | Vertical Elimination & Deeper first | Horizontal Elimination & Deeper first | Horizontal Elimination & 6 row first | Horizontal Elimination & 6 row center first

```python
import CandyCrashSimulationNXN_4 as S 
```

```python
def horizontal_or_vertical(board,r1,c1,d,height,width): # 0 mean horizontal; 1 mean vertical
    #print(np.array(board))
    x = [ 0, 0, 0,-1, 1] # 1:up 2:down 3:left 4:right
    y = [ 0,-1, 1, 0, 0]
    r = r1 + y[d]
    c = c1 + x[d]
    #print(r1,c1,d)
    #print(r,c)
    if c+1 < width and c-1 >= 0:
        #print('firstbbb')
        if (board[r1][c1] == board[r][c + 1] and board[r1][c1] == board[r][c - 1] and 
            not(r1 == r and c1 == c+1) and not(r1 == r and c1 == c-1)):
            #print('first')
            return 0
    if c+2 < width:
        #print('secondbbb')
        if (board[r1][c1] == board[r][c + 1] and board[r1][c1] == board[r][c + 2] and 
            not(r1 == r and c1 == c+1) and not(r1 == r and c1 == c+2)):
            #print('second')
            return 0
    if c-2 >= 0:
        #print('thirdbbb')
        if(board[r1][c1] == board[r][c - 1] and board[r1][c1] == board[r][c - 2] and 
            not(r1 == r and c1 == c-1) and not(r1 == r and c1 == c-2)):
            #print('third')
            return 0
    #print('vertial')
    return 1 
def Ncompare(num_move):
    x = [ 0, 0, 0,-1, 1] # 1:up 2:down 3:left 4:right
    y = [ 0,-1, 1, 0, 0]
    width = height = 9
    num_candy = 4
    eachscore = np.zeros(9) 
    eachscore[0] = num_move
    N = 1000
    for i in range(N):
        o_board = S.initialize(height,width)
        board = copy.deepcopy(o_board)
        #print("-----------------------------------------------------")
        for j in range(num_move): # random move
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            #print(np.array(board))
            ran = random.choice(move)
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[1] += score
            #print("after")
            #print(np.array(board))
            #print('random',ran,str(score))
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Vertical Elimination first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            vmove = [] 
            for i in move:                 # possible move
                if horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    vmove.append(i)
            if len(vmove) == 0:
                ran = random.choice(move)
            else:
                ran = random.choice(vmove)
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[2] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Horizontal Elimination first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            hmove = [] 
            for i in move:                 # possible move
                if not horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    hmove.append(i)
            if len(hmove) == 0:
                ran = random.choice(move)
            else:
                ran = random.choice(hmove)
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[3] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Vertical Elimination & Deeper first
            move = S.detect(board,height,width)
            
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            vmove = [] 
            for i in move:                 # possible move
                if horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    vmove.append(i)
            if len(vmove) == 0:
                ran = move[-1]
            else:
                vmove.sort()
                ran = vmove[-1]
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[4] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Horizontal Elimination & Deeper first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            hmove = [] 
            for i in move:                 # possible move
                if not horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    hmove.append(i)
            if len(hmove) == 0:
                ran = move[-1]
            else:
                hmove.sort()
                ran = hmove[-1]
                
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[5] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Horizontal Elimination & Deeper Center first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            hmove = [] 
            for i in move:                 # possible move
                if not horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    hmove.append(i)
            if len(hmove) == 0:
                move = sorted(move, key = lambda v: (v//100+y[v%10],abs((v%100)//10+x[v%10])-5))
                ran = move[-1]
            else:
                hmove = sorted(hmove, key = lambda v: (v//100+y[v%10],abs((v%100)//10+x[v%10])-5))
                ran = hmove[-1]
                
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[6] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Horizontal Elimination & 6 row first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            hmove = []
            move6row = []
            for i in move:                 # possible move
                if not horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    hmove.append(i)
                    if (i//100 == 6 and (i%10 == 3 or i%10 == 4)) or (i//100 == 5 and i%10 == 2) or (i//100 == 7 and i%10 == 1):
                        move6row.append(i)
            if len(move6row) != 0:
                ran = move6row[-1]
                #print(move6row)
            elif len(hmove) != 0:
                hmove.sort()
                ran = hmove[-1]
            else:
                ran = move[-1]
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[7] += score
        #print("-----------------------------------------------------")
        board = copy.deepcopy(o_board)
        for j in range(num_move): # Horizontal Elimination & 6 row center first
            move = S.detect(board,height,width)
            if len(move) == 0:
                board = S.initialize(height,width)
                move = S.detect(board,height,width)
            hmove = []
            move6row = []
            for i in move:                 # possible move
                if not horizontal_or_vertical(board,i//100,(i%100)//10,i%10,height,width):
                    hmove.append(i)
                    if (i//100 == 6 and (i%10 == 3 or i%10 == 4)) or (i//100 == 5 and i%10 == 2) or (i//100 == 7 and i%10 == 1):
                        move6row.append(i)
            if len(move6row) != 0:
                move6row = sorted(move6row, key = lambda v: (v//100+y[v%10],abs((v%100)//10+x[v%10])-5))
                ran = move6row[-1]
            elif len(hmove) != 0:
                hmove = sorted(hmove, key = lambda v: (v//100+y[v%10],abs((v%100)//10+x[v%10])-5))
                ran = hmove[-1]
            else:
                ran = move[-1]
            #print(ran)
            board = S.swap(board,ran//100,(ran%100)//10,ran%10)
            board,tmp,score = S.eliminate_to_static(board,height,width)
            eachscore[8] += score
        #print("-----------------------------------------------------")
    eachscore[1:] /= N
    return eachscore
```

```python
sresult = []
for i in range(1,2,1):
    tmp = Ncompare(i)
    sresult.append(tmp)
    print(tmp)

with open('compare9X9_8strategy.csv','w') as fout:
    writer = csv.writer(fout)
    writer.writerows(sresult)
```
